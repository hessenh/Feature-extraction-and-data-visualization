```python
from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100.)
```

```python
clf.fit(digits.data[:-1], digits.target[:-1])
```


Model Persistence

```python
from sklearn import svm
from sklearn import datasets
clf = svm.SVC()
iris = datasets.load_iris()
X, y = iris.data, iris.target
clf.fit(X, y)
```

```python
import pickle
s = pickle.dumps(clf)
clf2 = pickle.loads(s)
clf2.predict(X[0])
```
