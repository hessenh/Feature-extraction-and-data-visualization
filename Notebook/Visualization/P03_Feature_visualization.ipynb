<h1> Visualize different features </h1>

```python
import numpy as np
import pandas as pd
pd.set_option('html',False)

from os import path
```

<h3>Load the features </h3>

```python
P03_FEATURES = path.relpath('../data/P03/FEATURES/FEATURES.csv')
LABEL = path.relpath('../data/P03/DATA_WINDOW/Usability_LAB_All_L.csv')

df_features = pd.read_csv(P03_FEATURES, sep='\,')
df_label = pd.read_csv(LABEL, header=None, sep='\ ')
```

<h3>Group features by activity</h3>
1: Walking
2: Walking with transitions
3: Shuffeling
4: Stairs(acc)
5: Stairs(dec)
6: Standing
7: Sitting
8: Lying
9: Transition
10: Leaning
11: Undefined
12: Jumping
13: Dynamic
14: Static

```python
# Since df_features and df_label have different sizes, we have to change that
df_features = df_features[:len(df_label)-1]

#Extract what features we will look at
list_of_features = ['mean_thigh_x', 'mean_thigh_y', 'mean_thigh_z']
df_features = df_features[list_of_features]

# Walking: 
df_walking_features = df_features[df_label[0] == 1]
```

```python
import charts

options = dict(
    height=500, 
    title=dict(
        text='walking'),
    chart=dict(zoomType='xy')
)

charts.plot(df_walking_features, options=options, show='inline',  type='scatter')
```
