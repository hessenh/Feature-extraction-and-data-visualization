<h2> Visualize the raw data </h2>

```python
import numpy as np
import pandas as pd
pd.set_option('html',False)

from os import path
```

```python
from numpy import random
import matplotlib.pyplot as plt
import sys #only needed to determine Python version number

# Enable inline plotting
%matplotlib inline
#%matplotlib qt
#%pylab qt
SUB = "01A"

CHEST = path.relpath('../data/'+SUB+'/RAW_SIGNALS/'+SUB+'_Axivity_BACK_Back.csv')
THIGH = path.relpath('../data/'+SUB+'/RAW_SIGNALS/'+SUB+'_Axivity_THIGH_Right.csv')
LABEL = path.relpath('../data/'+SUB+'/RAW_SIGNALS/'+SUB+'_GoPro_LAB_All.csv')

df_chest = pd.read_csv(CHEST, header=None, sep='\,', engine='python')
df_thigh = pd.read_csv(THIGH, header=None, sep='\,', engine='python')
df_label = pd.read_csv(LABEL, header=None, sep='\ ', engine='python')


df_chest.iloc[0]
```

<h1>Testing new ploting</h1>

```python
import charts
```

```python
start = 0
length = 5000
df_chest = df_chest[start:start + length] + 15
df_thigh = df_thigh[start:start + length] + 20
df_label = df_label[start:start + length]
```

```python
#print df_label[df_label[0]==17]
#print df_label[0] == 17
```

```python
df_chest_thigh = pd.concat([df_chest, df_thigh, df_label],axis=1,ignore_index=True)

# Resample the data: Only take every 5th data point, to ease the plotting!
df_chest_thigh_resampled = df_chest_thigh.iloc[::5, :]

options = dict(
    height=1000, 
    title=dict(
        text='Chest and thigh'),
    chart=dict(zoomType='xy')
)

# Name for the plot. Saving the plot as START_STOP.svg. 
name = str(start) + "_" + str(length) + ".svg"
charts.plot(df_chest_thigh_resampled, options=options, save=name, show="inline")
```
