```python
CHEST_X = '/home/astrid/Development/Prosjektoppgave/Notebook/data/P03_DATA_WINDOW/PO3_CHEST_X.csv'
CHEST_Y = '/home/astrid/Development/Prosjektoppgave/Notebook/data/P03_DATA_WINDOW/PO3_CHEST_Y.csv'
CHEST_Z = '/home/astrid/Development/Prosjektoppgave/Notebook/data/P03_DATA_WINDOW/PO3_CHEST_Z.csv'

THIGH_X = '/home/astrid/Development/Prosjektoppgave/Notebook/data/P03_DATA_WINDOW/PO3_THIGH_X.csv'
THIGH_Y = '/home/astrid/Development/Prosjektoppgave/Notebook/data/P03_DATA_WINDOW/PO3_THIGH_Y.csv'
THIGH_Z = '/home/astrid/Development/Prosjektoppgave/Notebook/data/P03_DATA_WINDOW/PO3_THIGH_Z.csv'

LABEL = '/home/astrid/Development/Prosjektoppgave/Notebook/data/P03_DATA_WINDOW/P03_LABEL_L.csv'

df_chest_x = pd.read_csv(CHEST_X, header=None, sep='\,')
df_chest_y = pd.read_csv(CHEST_Y, header=None, sep='\,')
df_chest_z = pd.read_csv(CHEST_Z, header=None, sep='\,')

df_thigh_x = pd.read_csv(THIGH_X, header=None, sep='\,')
df_thigh_y = pd.read_csv(THIGH_Y, header=None, sep='\,')
df_thigh_z = pd.read_csv(THIGH_Z, header=None, sep='\,')

df_label = pd.read_csv(LABEL, header=None, sep='\ ')
```

```python
#bruker numpy fast fourier transform til Ã¥ transform et sliding window til frequency domain 

df_chest_250 = df_chest_x.iloc[250]

fft_output = np.fft.rfft(df_chest_x.iloc[250])

magnitude_only = [math.sqrt(i.real**2 + i.imag**2)/len(fft_output) for i in fft_output]
frequencies = [(i*1.0/100)*100 for i in range(100/2+1)]
```

```python
options = dict( 
    title=dict(
        text='time domain'),
    chart=dict(zoomType='xy')
)

charts.plot(df_chest_250, options=options, show='inline')
```

```python


options = dict(
    title=dict(
        text='frequency domain'),
    chart=dict(zoomType='xy')
)

frequencyArray = []
for i in range(0,50):
    elements = []
    elements.append(frequencies[i])
    elements.append(magnitude_only[i])
    frequencyArray.append(elements)

    
charts.plot(frequencyArray, options=options, show='inline')

#setup_graph(x_label='frequency (in Hz)', y_label='amplitude', title='frequency domain')
#plt.plot(frequencies, magnitude_only)
#plt.show()
```
