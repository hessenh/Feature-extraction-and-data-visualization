```python
def fourierTransform(slidingWindow):
    #sliding window from time domain to frequency domain
    fft_output = np.fft.fft(slidingWindow)
    magnitude = abs(fft_output)
    frequencies = np.fft.fftfreq(100)
    #magnitude_only = [math.sqrt(i.real**2 + i.imag**2)/len(fft_output) for i in fft_output]
    #frequencies = [(i*1.0/100)*100 for i in range(100/2+1)]
    
    #Make plotable array
    plotArray = []
    for i in range(0,50):
        elements = []
        elements.append(frequencies[i])
        elements.append(magnitude[i])
        plotArray.append(elements)
    
    return fft_output, magnitude, frequencies, plotArray
    
```

```python
options = dict( 
    title=dict(
        text='time domain'),
    chart=dict(zoomType='xy')
)

charts.plot(df_chest_x.iloc[100], options=options, show='inline')
```

```python
fft_output, magnitude, frequencies, plotableArray = fourierTransform(df_chest_x.iloc[360])



options = dict(
    title=dict(
        text='frequency domain'),
    chart=dict(zoomType='xy')
)

charts.plot(plotableArray, options=options, show='inline')
#print fft_output
#print magnitude
```
